```python
# Operador == 

Livro_1 = input("Digite o titulo do  1 livro: ")
Livro_2 = input("Digite o titulo do segundo livro")

if  Livro_1 == Livro_2:
  print("Os livros tem o mesmo titulo")
```

```python
media  = float(input("Digite sua media:"))

if media >= 6.0:
  print("Aprovado(a)")
else:
  print("Reprovado(a)")
```

```python
# Para os alunos que tenham a media entre 4.0 e 6.0

media = float(input("Digite a media: "))

if media >= 6.0:
  print("Aprovado")
if 6.0 > media >= 4.0:
  print("Recuperacao")
if media < 4.0:
  print("Reprovado:")
```
