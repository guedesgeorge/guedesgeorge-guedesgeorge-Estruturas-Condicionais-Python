```python
# 1. Entrada de dados
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))
operacao = input("Qual operação deseja realizar? (+, -, *, /): ")

# 2. Realizando o cálculo e salvando o resultado
if operacao == "+":
    resultado = num1 + num2
elif operacao == "-":
    resultado = num1 - num2
elif operacao == "*":
    resultado = num1 * num2
elif operacao == "/":
    if num2 != 0: # Travinha de segurança para não dividir por zero
        resultado = num1 / num2
    else:
        resultado = None
        print("Erro: Não é possível dividir por zero.")
else:
    resultado = None
    print("Operação inválida!")

# 3. Fazendo a identificação (apenas se houver um resultado válido)
if resultado is not None:
    # Identificar Positivo ou Negativo
    if resultado > 0:
        sinal = "positivo"
    elif resultado < 0:
        sinal = "negativo"
    else:
        sinal = "neutro (zero)"

    # Identificar Inteiro ou Decimal
    if resultado % 1 == 0:
        tipo = "inteiro"
    else:
        tipo = "decimal"

    # Identificar Par ou Ímpar (geralmente aplicado a inteiros)
    if resultado % 2 == 0:
        paridade = "par"
    else:
        paridade = "ímpar"

    # 4. Exibindo o resultado final
    print("-" * 30)
    print(f"Resultado da operação: {resultado}")
    print(f"O número é {sinal}, {tipo} e {paridade}.")
```
